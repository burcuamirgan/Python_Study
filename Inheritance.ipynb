# OOP...
Object Oriented Programing (Nesne Tabanlı Programlama) ile çeşitli nesneler oluşturularak complex yazılımlar kolayca oluşturulabilir.
Class (sınıf) yapısı, attribute (özellik) ve function (metot) birleşimden oluşmaktadır.

Personal_Info adında bir class oluşturalım.
Bu class kişisel bilgileri içeren attribute yapısına (isim,yaş,boy,kilo,meslek) sahip olsun. 

```python
class Personal_Info():
    name = "Ahmet"
    age = 13
    weight = 40
    job = "öğrenci"
```

```python
print("Created class structure:{}".format(Personal_Info))
```

```python
new_person = Personal_Info()
```

```python
print("Function and attributes for new_person:\n{}".format(dir(new_person)))
```

```python
print("Some attributes: \nName:{} \nAge:{}".format(new_person.name,new_person.age))
```

Bu class için kullanılan attribute değerlerinin farklı değerler ile oluşturulabilmesi için __init__() metodu kullanılır.
init metodu, class oluşturulurken ilk çağrılan metodtur.
Self'de classı gösteren bir referans ve self.attribute olarak kullanılır.

```python
class Personal_Info():
    def __init__(self,name,age,weight,job):
        self.name = name
        self.age = age
        self.weight = weight
        self.job = job
```

```python
new_person1 = Personal_Info("Ahmet",13,40,"öğrenci")
new_person2 = Personal_Info("Ayşe",25,55,"öğretmen")
print("Person1 Name:{} \nPerson2 Name:{}".format(new_person1.name,new_person2.name))
```

# Inheritance (Kalıtım/Miras)
Inheritance, herhangi bir sınıfın özellik ve metotlarının bir kısmının yada tamamının başka bir sınıfa miras olarak aktarılmasını sağlar.

```python
class Job_Personal_Info(Personal_Info): # Personal_Info sınıfı miras alınır.
    pass
#     def __init__(self,salary,experience):
#         self.salary = salary
#         self.experience = experience
```

```python
job_person = Job_Personal_Info("Ayşe",25,55,"öğretmen")
```

```python
print("Function and attributes for Job_Personal_Info:\n{}".format(dir(job_person)))
```

```python
class Job_Personal_Info2(Personal_Info): # Personal_Info sınıfı miras alınır.
#     pass
    def soyad(self, soy_adi): # Miras aldığımız Personal_Info sınıfına soyad adlı yeni fonk ekledik.
        self.name += soy_adi
```

```python
job_person_surname = Job_Personal_Info2("Veli",40,90,"Bankaci")
```

```python
job_person_surname.soyad("Başoğlu")
```

```python
job_person_surname.name
```

# Overriding (İptal Etme)


Miras alınan bir fonksiyonun bu class'ta yeniden oluşturulması ile eski fonksiyon override (iptal) edilmiş olur ve yeni oluşturulan fonksiyon geçerli olur.

```python
print("Function and attributes for Job_Personal_Info2:\n{}".format(dir(Job_Personal_Info2)))
```

```python
class Job_Personal_Info3(Job_Personal_Info2):
    def soyad(self, soy_adi): # Miras aldığımız Personal_Info sınıfına soyad adlı yeni fonk ekledik.
        self.name += " {}".format(soy_adi)
```

```python
job_person_name_surname = Job_Personal_Info3("Merve",28,50,"Mühendis")
```

```python
job_person_name_surname.soyad("Türk")
```

```python
job_person_name_surname.name
```

# .super keyword

Miras aldığımız attribute'ları alt sınıflarda kullanmamızı sağlar.

```python
class Personal_Info():
    def __init__(self,name,age,weight,job):
        self.name = name
        self.age = age
        self.weight = weight
        self.job = job
```

```python
class Çalışan_profili(Personal_Info):
    def __init__(self,name,age,weight,job,salary):
        super().__init__(name,age,weight,job) # 3 tane özelliği Personal_Info fonksiyonunun init fonksiyonuyla alıyoruz.
        print("Çalışan profili oluşturuluyor...")
        self.salary = salary # job özelliği de kendimiz yazıyoruz.
    def worker_info(self):
        print("Çalışan profilleri gösteriliyor...")
        print("İsim: {} \nYaş: {} \nKilo: {} \nBölüm: {} \nMaaş: {}".format(self.name,self.age,self.weight,self.job,self.salary))
```

```python
worker = Çalışan_profili("Mehmet Ali",35,85,"Muhasebe",6000)
worker = Çalışan_profili("Mehmet Ali2",35,85,"Muhasebe",6000)
```

```python
worker.worker_info()
```
